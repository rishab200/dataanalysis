```python
def rating(rating):
  if rating>=8.5:
    return 'Excellent'
  elif rating>=7.5:
    return 'Good'
  elif rating>=6.5:
    return 'Average'
  else:
    return 'Bad'
```

```python
df["Rating_cat"]=df['Rating'].apply(rating)
```
