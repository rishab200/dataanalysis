# **locations of restaurants on a map**

```python
import geopandas as gpd
import plotly.express as px
```

```python
df_geo=gpd.GeoDataFrame(df,geometry=gpd.points_from_xy(df.Longitude,df.Latitude))
```

```python
plt.figure(figsize=(20,16))
world=gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ax=world.plot(color='lightgreen',edgecolor='white')
df_geo.plot(ax=ax,color='orange',markersize=3)
plt.title("Locations of Restaurants on a Map")
plt.show()
```

```python
import folium
from folium.plugins import HeatMap
```

```python
world_map=folium.Map(location=[df['Latitude'].mean(),df['Longitude'].mean()],zoom_start=2,height=600,width=800)
```

```python
heatmarker=[[row['Latitude'],row['Longitude']] for index,row in df.iterrows()]
HeatMap(heatmarker,radius=10).add_to(world_map)
```

# **restaurant chains**

```python
reastaurant_chain=df.groupby('Restaurant Name')['Restaurant Name'].count().reset_index(name='Outlets')
```

```python
reastaurant_chains=reastaurant_chain.sort_values(by='Outlets',ascending=False).head(4)
```

```python
reastaurant_chains[['Restaurant Name','Outlets']]
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=reastaurant_chains['Restaurant Name'],y=reastaurant_chains['Outlets'],palette='Set2')
plt.title("Restaurant Chains")
plt.show()
```

```python
rating=df.groupby('Restaurant Name')['Aggregate rating'].mean().reset_index(name='Rating')['Rating']
```

```python
rating
```

```python
sns.barplot(x=reastaurant_chains['Restaurant Name'],y=rating,palette='Set2')
plt.title("Restaurant Chains")
plt.show()
```

```python
popularity=df.groupby('Restaurant Name')['Votes'].sum().reset_index(name='Votes')['Votes']
```

```python
popularity
```

```python
sns.barplot(x=reastaurant_chains['Restaurant Name'],y=popularity,palette='Set2')
plt.title("Restaurant Chains")
plt.show()
```
